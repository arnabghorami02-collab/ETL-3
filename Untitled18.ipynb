# Handling Missing Data in ETL Assignment

Q1. What are the most common reasons for missing data in ETL pipelines?

Sol :      

Common reasons are:

- Manual data entry errors – users forget to fill fields.
- System integration issues – data is lost while transferring between systems.
- Sensor or device failure – machines fail to capture values.
- Database migration problems – some values are not copied correctly.
- Different data formats – incompatible formats create null values.
- Optional fields – some fields are not mandatory.
- Network or API failures – incomplete records are received.
- Data corruption – damaged files or records during processing.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

Sol :       

Blind deletion is bad because:

- It can remove a large amount of useful data.
- Important business information may be lost.
- Dataset size becomes smaller, reducing analysis accuracy.
- It may create biased results if missing data follows a pattern.
- Some missing values themselves may provide insights.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

Sol :       
Median is preferred because:

- Income data usually contains extreme values (outliers).
- Mean gets heavily affected by very high or very low values.
- Median represents the middle value and is more stable.
- Therefore, median gives a more realistic replacement.

Q5. What is forward fill and in what type of dataset is it most useful?

Sol :      
Forward fill means replacing missing values with the previous available value.

Example:

It is most useful in:

- Time-series data
- Sales records
- Stock prices
- Sensor readings

Q6. Why should flagging missing values be done before imputation in an ETL workflow?

Sol :    
Because after imputation, original missing information disappears.

Flagging helps:

- Track which values were originally missing
- Improve transparency
- Help machine learning models detect patterns
- Preserve business insights

Q7. Consider a scenario where income is missing for many customers.
 How can this missingness itself provide business insights?

 Sol :     
Missing income may indicate:

- Customers unwilling to share salary details
- Poor form design
- Certain regions/customers having incomplete records
